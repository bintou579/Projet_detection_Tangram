```
item_list = {"Maison": 0,
             "Bateau": 60+25,
             "Pont": 4*60+4,
             "Cygne": 6*60,
             "Chat": 7*60+5,
             "Bol": 8*60+2,
             "Marteau": 9*60+40,
             "Coeur": 10*60,
             "Lapin": 11*60+31,
             "Renard": 13*60+22}

print(item_list)
```

```
#Création de la dataframe
data = {'Imagelist': os.listdir('trigram/vid2im'),
        'Classname': ['']* len(os.listdir('trigram/vid2im'))}

for pic in data['Imagelist']:
  print(pic)
```

```
data['Imagelist'].index('image533.jpg')
```

Comme les images sont dans le jeu de données de manière complètement aléatoire, on va itérer pour chacune d'entre elles.

```
current_label = "Maison"
t = 0
is_above = [ts <0 for ts in item_list.values() ]
framerate = 0.5
for i in range(1, len(data['Imagelist'])+1):
  data['Classname'][data['Imagelist'].index(f'image{i}.jpg')] = current_label
  t += 0.5
  if [ts <t for ts in item_list.values() ] != is_above:
    is_above = [ts <t for ts in item_list.values() ]
    current_label = list(item_list.keys())[sum(is_above)-1]
```

```
df = pd.DataFrame(data)
df.head()
```

```
#Optionnel : on peut marquer des images pour le train et le test.
df['Split'] = ['val'] * df.shape[0]
df.loc[:int(df.shape[0]*0.8),'Split'] = 'train'
df.head()
```

```
df['new_image_name'] = df['Imagelist']

pic_count = {"Maison": 0,
             "Bateau": 0,
             "Pont": 0,
             "Cygne": 0,
             "Chat": 0,
             "Bol": 0,
             "Marteau": 0,
             "Coeur": 0,
             "Lapin": 0,
             "Renard": 0}


for i in range(df.shape[0]):
  df['new_image_name'][i] = f"{df['Classname'][i].lower()}.{pic_count[df['Classname'][i]]}.jpg" 
  pic_count[df['Classname'][i]] += 1

df.head()
```

```
#On sauvegarde la dataframe
df.to_csv('/mydrive/trigram_labels.csv')
```
