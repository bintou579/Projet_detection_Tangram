```
import os
import cv2
```

```
count = 0

for directory in os.listdir(path):
  new_path = os.path.join(path, directory)
  for filename in os.listdir(new_path):
    img_path = os.path.join(new_path, filename)
    img = cv2.imread(img_path)
    height, width, channels = img.shape
    count = count + 1
    # Cut the image in half
    width_cutoff = width // 2
    s1 = img[:, :width_cutoff]
    s2 = img[:, width_cutoff:]

    # Save each half
    cv2.imwrite(new_path + "/split1_%d.jpg" % count, s1)
    cv2.imwrite(new_path + "/split2_%d.jpg" % count, s2)
```

```
from PIL import Image, ImageEnhance
```

```
for directory in os.listdir(path):
  new_path = os.path.join(path, directory)
  count = 0
  for filename in os.listdir(new_path):
    if str(filename).startswith("split"):
      count = count + 1
      img_path = os.path.join(new_path, filename)
      img = Image.open(img_path)
      #image brightness enhancer
      enhancer = ImageEnhance.Brightness(img)

      factor = 1 #gives original image
      im_output = enhancer.enhance(factor)
      im_output.save(path+"/"+str(directory)+"/"+str(directory)+"_%d.jpg" % count)

      factor = 0.5 #darkens the image
      im_output = enhancer.enhance(factor)
      im_output.save(path+"/"+str(directory)+"/"+str(directory)+"_dark_%d.jpg" % count)

      factor = 1.5 #brightens the image
      im_output = enhancer.enhance(factor)
      im_output.save(path+"/"+str(directory)+"/"+str(directory)+"_clear_%d.jpg" % count)
    else:
      pass
```
